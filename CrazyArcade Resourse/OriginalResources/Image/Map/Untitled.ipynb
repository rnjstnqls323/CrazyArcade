```python
import os
from PIL import Image

folder_path = '.'  # 변환할 이미지가 있는 폴더

target_color = (255, 0, 255)  # 투명하게 바꿀 핑크색 RGB

for filename in os.listdir(folder_path):
    if filename.lower().endswith('.bmp'):
        bmp_path = os.path.join(folder_path, filename)
        png_filename = filename.rsplit('.', 1)[0] + '.png'
        png_path = os.path.join(folder_path, png_filename)

        try:
            img = Image.open(bmp_path).convert("RGBA")  # RGBA로 변환

            datas = img.getdata()
            new_data = []

            for item in datas:
                # RGB 값이 target_color(핑크)이면 알파 0 (투명)
                if item[:3] == target_color:
                    new_data.append((255, 255, 255, 0))  # 투명 처리
                else:
                    new_data.append(item)

            img.putdata(new_data)
            img.save(png_path)
            print(f"변환 및 투명 처리 완료: {filename} → {png_filename}")
        except Exception as e:
            print(f"변환 실패: {filename} ({e})")
```

```python
import os
from PIL import Image

folder_path = '.'  # 이미지가 있는 폴더
target_color = (255, 0, 255)  # 투명하게 바꿀 핑크색 RGB

def is_target_color(pixel):
    return pixel[:3] == target_color

for filename in os.listdir(folder_path):
    if filename.lower().endswith('.png'):
        png_path = os.path.join(folder_path, filename)

        try:
            img = Image.open(png_path).convert("RGBA")
            datas = img.load()  # 픽셀 접근용

            width, height = img.size

            # 이미지 가장자리 픽셀 검사
            # 가장자리 픽셀 중 핑크색이면 투명 처리
            for x in range(width):
                # 위쪽 테두리
                if is_target_color(datas[x, 0]):
                    datas[x, 0] = (255, 255, 255, 0)
                # 아래쪽 테두리
                if is_target_color(datas[x, height - 1]):
                    datas[x, height - 1] = (255, 255, 255, 0)

            for y in range(height):
                # 왼쪽 테두리
                if is_target_color(datas[0, y]):
                    datas[0, y] = (255, 255, 255, 0)
                # 오른쪽 테두리
                if is_target_color(datas[width - 1, y]):
                    datas[width - 1, y] = (255, 255, 255, 0)

            # 저장 (덮어쓰기)
            img.save(png_path)
            print(f"배경 핑크색 투명 처리 완료: {filename}")

        except Exception as e:
            print(f"변환 실패: {filename} ({e})")
```
